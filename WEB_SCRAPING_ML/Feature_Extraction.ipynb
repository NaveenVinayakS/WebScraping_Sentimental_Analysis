```python
import pandas as pd
df = pd.read_csv('scraping.csv')
df.head()
```

#### Number Of words

```python
df['word_count'] = df['Comment'].apply(lambda x: len(str(x).split(" ")))
df[['Comment','word_count','Rating']].head()
```

####  Number of characters

```python
df['char_count'] = df['Comment'].str.len() ## this also includes spaces
df[['Comment','char_count','Rating']].head()
```

#### Average length of Word

```python
def avg_word(sentence):
    words = sentence.split()
    return (sum(len(word) for word in words)/len(words))

df['avg_word'] = df['Comment'].apply(lambda x: avg_word(x))
df[['Comment','avg_word','Rating']].head()
```

#### Number of stopwords

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

df['stopwords'] = df['Comment'].apply(lambda x: len([x for x in x.split() if x in stop]))
df[['Comment','stopwords','Rating']].head()
```

#### Number of special characters

```python
df['hastags'] = df['Comment'].apply(lambda x: len([x for x in x.split() if x.startswith('#')]))
df[['Comment','hastags','Rating']].head()
```

#### Number of numerics

```python
df['numerics'] = df['Comment'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df[['Comment','numerics','Rating']].head()
```

#### Number of Uppercase words

```python
df['upper'] = df['Comment'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df[['Comment','upper','Rating']].head()
```
